```python
from bs4 import BeautifulSoup
import requests
import csv
```

```python
# create csv file
outFile = open('bdjobsJobs_scrap.csv', 'w', encoding='utf8', newline='')
writer = csv.writer(outFile)
writer.writerow(['Job Ttile','Company Name','Job Location','Required Experience','Salary' ,'Job Details Link'])
```

```python
url = 'https://jobs.bdjobs.com/jobsearch.asp?fcatId=8&icatId='
# url = 'https://jobs.bdjobs.com/jobsearch.asp?fcatId=8'
r = requests.get(url)
soup = BeautifulSoup(r.content, 'html.parser')
```

```python
def getSalary(link):
    _r = requests.get("https://jobs.bdjobs.com/" +link)
    _soup = BeautifulSoup(_r.content, 'lxml')
    _jobs = _soup.find('div', class_='salary_range')
    return _jobs
```

```python
jobs = soup.find_all('div', class_=['norm-jobs-wrapper', 'sout-jobs-wrapper'])

for index,job in enumerate(jobs, 1):
    try:
        jt = job.find('div', class_='job-title-text')

        job_title = jt.a.text.replace('\r\n','').replace(' ','')
        job_link = jt.a['href']
        
        company_name = job.find('div', class_='comp-name-text').text.replace('\n','').replace('\r','').replace(' ','')

        job_loc = job.find('div', class_='locon-text-d')
        job_location = job_loc.text.replace('\n','')

        job_exp = job.find('div', class_='exp-text-d')
        job_experience = job_exp.text.replace('\n','').replace('\r','').replace(' ','')

        job_deadline = job.find('div', class_='dead-text-d').text.replace('\r\n','').replace('\n','')
        
        salary = getSalary(job_link).text.replace('\r\n', '').replace('\n', '').replace(' ', '').replace('Salary', '')

    except Exception:
        print(Exception)
   
    print(f'''{index}
    Job Title: {job_title}
    Company Name: {company_name}
    Job Location: {job_location}
    Required Experience: {job_experience}
    Job Deatils: {"https://jobs.bdjobs.com/"+job_link}
    Salary: {salary}
    ''')
    
    writer.writerow([job_title, company_name, job_location, job_experience, salary ,"https://jobs.bdjobs.com/"+job_link])

outFile.close()
```
